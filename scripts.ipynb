#### Text Wrap

```


def wrap(string, max_width):
    i=0
    result=''
    while(i<len(string)+1):
        result = result + string[i:i+max_width]+'\n'
        i+=max_width
    return(result)

```

#### sWAP cASE

```
def swap_case(s):
    
    output=''
    #ord() function returns ASCII code of chr
    #chr() function returns char corresponding to ASCII code
    #ord('a') == ord('A')+32
    for char in s:
        if(ord(char)>=ord('a') and ord(char)<=ord('z')):
            output+=chr(ord(char)-32) 
            
        elif(ord(char)>=ord('A') and ord(char)<=ord('Z')):
            output+=chr(ord(char)+32) 
        else:
            output+=char
    return(output)

```

#### HTML Parser - Part 1

```
# Enter your code here. Read input from STDIN. Print output to STDOUTimport re
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print ("Start :", tag)
        self.value(attrs)

    def handle_endtag(self, tag):
        print ("End   :", tag)

    def handle_startendtag(self, tag, attrs):
        print ("Empty :", tag)
        self.value(attrs)

    def value(self, attrs = None):
        if attrs:
            for attr,val, in attrs:
                print('->', attr, '>', val)


            
n=int(input())
ss = '\n'.join([input() for i in range(n)])#cycle found on solutions, does not work if built inside parenthesis
parser = MyHTMLParser()
parser.feed(ss)
```

#### HTML Parser - Part 2

```
# Enter your code here. Read input from STDIN. Print output to STDOUT

from html.parser import HTMLParser


class MyHTMLParser(HTMLParser):
    
    def handle_comment(self, data):
        
        if "\n" in data:
            print(">>> Multi-line Comment")
            print(data)
        else:
            print(">>> Single-line Comment")
            print(data)
    
    def handle_data(self, data):
        
        if data != "\n":
            print(">>> Data")
            print(data)
             
html = ""
n=int(input())
for i in range(n):
    html += input().rstrip()
    html += '\n'
     
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```
